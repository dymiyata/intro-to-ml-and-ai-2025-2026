```python
def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  left = merge_sort(arr[:mid])
  right = merge_sort(arr[mid:])
  return merge(left, right)

def merge(arr1, arr2):
  result = []
  i = 0 # our position in arr1
  j = 0 # our position in arr2

  while i < len(arr1) and j < len(arr2):  # both lists still have stuff to add
    if arr1[i] < arr2[j]:
      result.append(arr1[i])
      i += 1
    else:
      result.append(arr2[j])
      j += 1

  while i < len(arr1): # still stuff in arr1
    result.append(arr1[i])
    i += 1

  while j < len(arr2): # still stuff in arr2
    result.append(arr2[j])
    j += 1

  return result
```

```python
merge_sort([7, 11, 6, 1, -5, 8, 13, 4, 3.14])
```
