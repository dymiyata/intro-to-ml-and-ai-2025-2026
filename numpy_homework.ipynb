In the following cell, I've defined all of these for you in code.

```python
### Don't Edit this cell at all
### Create new cells to run your own code

# Feature Matrix
X = np.array([
    [2, 5, 1],
    [3, 2, 4],
    [1, 3, 2],
    [4, 1, 3],
    [5, 2, 1],
    [2, 4, 5],
    [3, 3, 3]
])

# Target vector
Y = np.array([10, 12, 7, 11, 9, 14, 10])

# Weight vector
w = np.array([1.2, 1, 1.5])

# Bias
b = 0.2
```
