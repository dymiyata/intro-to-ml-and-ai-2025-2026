# Implementing a Random Forest Classifier

First we import some of the usual stuff.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Dataset

For this example, we will use seaborn's built in `"titanic"` dataset. This is the same dataset we used in when we first implemented a decision tree.

Additionally, we will only use some of the features (the same ones as before) and our target variable will be `"survived"` (whether or not the given person survived).

```python
df = sns.load_dataset("titanic")
df_filtered = df[["sex", "pclass", "age", "fare", "survived"]].dropna()
```

Just as before, we define our feature matrix `X` and target vector `y` using this data.

```python
# X = df_filtered[["sex", "pclass", "age", "fare"]]
# y = df_filtered["survived"]
X = df.drop(columns=["survived", "alive"])
y = df["survived"]
```

```python
X.head()
```

Before we continue, recall that for decision trees we needed to convert categorical variables with many values into many binary variables. We do this with one-hot enconding.

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

Now that we've done the preprocessing, we can split our data into a training and testing set.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=.2,
    random_state=2026,
    stratify=y
)
```

To compare a random forest to a single decision tree, we'll run both.  The Random Forest model is in the `sklearn.ensemble` module. (whereas the decision tree model is in the `sklearn.tree` module)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

We define the models.

```python
tree_model = DecisionTreeClassifier()
forest_model = RandomForestClassifier()
```

Train the models on the training data:

```python
tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)
```

We can evaluate the models with our metrics.

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred_train_tree = tree_model.predict(X_train)
y_pred_test_tree = tree_model.predict(X_test)
y_pred_train_forest = forest_model.predict(X_train)
y_pred_test_forest = forest_model.predict(X_test)
```

```python
print(classification_report(y_train, y_pred_train_tree))
print(classification_report(y_test, y_pred_test_tree))
```

```python
print(classification_report(y_train, y_pred_train_forest))
print(classification_report(y_test, y_pred_test_forest))
```

```python
tree_model = DecisionTreeClassifier()
forest_model = RandomForestClassifier(random_state=2026)

tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)

y_pred_train_tree = tree_model.predict(X_train)
y_pred_test_tree = tree_model.predict(X_test)
y_pred_train_forest = forest_model.predict(X_train)
y_pred_test_forest = forest_model.predict(X_test)

print("Tree Training Set:")
print(classification_report(y_train, y_pred_train_tree))
print("Tree Testing Set:")
print(classification_report(y_test, y_pred_test_tree))


print("Forest Training Set:")
print(classification_report(y_train, y_pred_train_forest))
print("Forest Testing Set:")
print(classification_report(y_test, y_pred_test_forest))
```
